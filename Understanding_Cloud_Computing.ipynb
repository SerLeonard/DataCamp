# **Introduction to Cloud Computing**

## **Cloud Computing**

*  cloud computing is the delivery of technology services -
*  including compute, storage, databases, networking, software, and many more -
*  over the internet with pay-as-you-go pricing


### **Hosting a website using an on-premise serve**

*  Server
   *  buying a server and hosting the website on that server
   *  powerful computer
   *  accessible remotely- hosted on-site
*  located on the premises  

When traffic increases  

*  more people start using the website
*  buy/rent new servers
   *  take time to setp up
   *  cost a lot of money


### **Hosting a website using a cloud server**

*  cloud server
*  acccess to computing power instantly when you need it

When traffic increases  

*  more people start using the website
*  access more cloud servers
*  easily release redundant cloud servers
*  pay-as-you-go billing

## **Cloud Service Models**

*  **On-premise**
   *  responsible for everything
*  **IaaS** (Infrastructure as a Service)
   *  "rent" the fundamental building blocks-
   *  networking, storage, and servers
*  **PaaS** (Platform as a Service)
   *  includes the underlying infrastructure plus tools and software needed to build applications
   *  operating systems and middleware
*  **SaaS** (Software as a Serice)
   *  provides software via web browser or app, with the hardware and OS manages by the vendor (cloud providers)


#### **IaaS**

*  cloud-based alternative to on-premise infrasture
*  **advantages:**
   *  scalable alternative to expensive on-premise infrastructure
*  **users:**
   *  system admins
*  **examples:**
   *  cloud servers from e.g., google compute engine, microsoft azure, amazon web services

#### **PaaS**

*  hardware and software tools over the internet used to develop applications
*  **advantages:**
   *  developers don't need to start from teh scratch when creating applications
*  **users:**
   *  developers
*  **examples:**
   *  web apps, logic apps e.g., google app engine, windows azure, AWS elastic beanstalk

#### **SaaS**

*  software available over the internet, usually for a monthly subscription fee
*  **advantages:**
   *  no need to install software on you computer
*  **users:**
   *  end customers
*  **examples:**
   *  internet applications e.g., google g suite, microsoft office 365, dropbox


### **The cloud pyramid**

top  
<----------------------- Abstraction  
SaaS      Paas      Iaas  
------------------------> control  
bottom

*  as we move up, abstraction increases and control decreases
*  more abstraction means less details but easier management
*  the highest level of abstraction covers the entire system while the lower levels involve fewer components


### **Other cloud service models**

   *  uses a "serverless" billing model

# **Cloud Deployment**

## **Cloud deployment models**

*  important decision in cloud adoption
*  how much control do you need over your cloud environemnt?
*  three main types: **private, public, and hybrid**


### **Private cloud**

*  Cloud intrastructure is designated for exclusive use by its tenants  

*  Private clouds are accessed by a network link  

*  **Pros:** direct control of resources and data  

*  **Cons:** more upfront investment

*  unlike on-premise, private cloud uses virtualization for on-demand compute resources and can be off-premises


### **Public cloud**

*  cloud infrastructure is shared and open for use by the general public. It's owned managed by a cloud service provider

*  public clouds are internet accessible

*  **Pros:**
   *  get started quickly with minimal investment
   *  easier to scale

*  **Cons:**
   *  no access to data center and hardware


### **Hybrid cloud**

*  organization uses a combination of two or more district models:  

**Use case:**

*  store sensitive data on the private cloud and use application on public cloud or analytics

*  **cloud bursting**
   *  when private cloud hits capacity, temporarily move overflow to the public cloud to avoid disruption of service


### **Other deployment models**

*  **Multicloud**
   *  combination of different cloud provider services
   *  flexibility on pricing plans and service offerings
   *  no reliance on one vendor (cloud provider)

*  **Community**
   *  infratructure shared by a specific community for exclusive use
   *  common interest or concern, e.g., security, jurisdiction, mission
   *  can be managed and hosted internally or externally

## **Regulations on the cloud**

*  data centers/servers are spread out across the globe reduces latency

*  the greater the distance betweem two points, the longer it will take for the data to get there

### **Latency**

*  the delay between the moment data is transmitted and the moment it is received is known as **latency**


### **General Data Protection Regulation (GDPR)**

*  regulates how personal data is collected, processed, and stored from used in the EU

*  **Examples:**
   *  users must explicitly consent to data collection
   *  notify users of any data breaches
   *  personal data information must be encypted, anonymized, and/or pseudonymized
   *  **personal data can't leave EU borders, unless you can guarantee the same level of protection**

*  **Fine:** 20 million Euros or up to 4% of the worldwide annual revenue



### **Personal data**

*  Personal data is any information that relates to an identified or identifiable living individual.  

*  Different pieces of information, which collected together can lead to the identification of a particular person, also constitute personal data

*  includes: home address, first name, last name, email address, location data, IP address, racial or ethnic origin, political opinions, sexual orientation, health related data

*  personal identifiable information (PII)
